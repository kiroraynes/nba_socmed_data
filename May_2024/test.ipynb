```python
import requests
ids = [elem.replace("https://www.youtube.com/channel/","") for elem in df['Youtube'].values]
yt_followers = []
batch_size = 50
for start in range (0, len(df['Youtube']), batch_size):
    string = ",".join(ids[start: start+batch_size])
    url = "https://www.googleapis.com/youtube/v3/channels"
    params = {
        "part": "statistics",
        "id": string,
        "key": ""
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        file = response.json()
        for i in file['items']:
            index_number = df.index[df['Youtube'] == ("https://www.youtube.com/channel/" + str(i['id']))].tolist()
            df.at[index_number[0],'YouTube (subscribers)'] = i['statistics']['subscriberCount']

df.to_excel("may_2024.xlsx", index=False)
```
